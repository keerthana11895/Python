```python
#Check if amount is Positive, Negative, or Zero
amount = int(input("Enter amount: "))

if amount > 0:
    print("Positive amount")
elif amount < 0:
    print("Negative amount")
else:
    print("Zero amount")
```

```python
#Sum of digits of a number
num = int(input("Enter passcode: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num //= 10

print("Sum of digits:", total)
```

```python
#Reverse a number
num = int(input("Enter number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

```python
#Check if a number is Prime
num = int(input("Enter user ID: "))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
#Factorial using Recursion
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
print("Factorial:", factorial(num))
```

```python
#Check Armstrong Number
num = int(input("Enter number: "))
temp = num
digits = len(str(num))
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** digits
    temp //= 10

if total == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
#Swap first and last character of a string
s = input("Enter password: ")

if len(s) <= 1:
    print(s)
else:
    result = s[-1] + s[1:-1] + s[0]
    print(result)
```

```python
#Decimal to Binary conversion
num = int(input("Enter decimal number: "))
binary = ""

while num > 0:
    binary = str(num % 2) + binary
    num //= 2

print("Binary:", binary)
```

```python
#Find the longest word in a sentence
sentence = input("Enter sentence: ")

words = sentence.split()
longest = max(words, key=len)

print("Longest word:", longest)
```

```python
#Check if two strings are Anagrams
s1 = input("Enter first string: ").replace(" ", "").lower()
s2 = input("Enter second string: ").replace(" ", "").lower()

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not anagrams")
```
