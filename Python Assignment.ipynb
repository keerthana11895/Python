```python
#7. Calculate final amount after 20% discount for orders > $100
```

```python
total = float(input("Enter total amount: "))
```

```python
if total > 100:
    discount = total * 0.20
else:
    discount = 0

final_amount = total - discount
print("Final amount:", final_amount)
```

```python
#8. Check withdrawal balance and update remaining balance
```

```python
balance = float(input("Enter your balance: "))
```

```python
withdraw_amount = float(input("Enter amount to withdraw: "))
```

```python
if withdraw_amount <= balance:
    balance -= withdraw_amount
    print("Withdrawal successful.")
    print("Remaining balance:", balance)
else:
    print("Insufficient balance.")
```

```python
#9. Check if a year is a leap year
```

```python
year = int(input("Enter year: "))
```

```python
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
#10. Return only even numbers from a list
```

```python
numbers = [10, 15, 22, 37, 40, 51]
```

```python
even_numbers = [n for n in numbers if n % 2 == 0]

print("Even numbers:", even_numbers)
```
